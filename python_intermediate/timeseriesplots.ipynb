```python
import numpy as np
import pandas as pd
np.random.seed(50)

def random_walk(n_steps):
    step= 0.5 - np.random.random(n_steps)
    return np.cumsum(step)

## date_range() generates a equality spaced time range
date =pd.date_range('1/1/2019', '12/31/2019', freq='D')
data = {'a':random_walk(365), 'b':random_walk(365)}
sales_data=pd.DataFrame(data,index=date)
sales_data ## dataframe with 365 days and two variables
```

```python
import pandas as pd
```

```python
appledf = pd.read_csv('apple-historical-data.csv', parse_dates = ['Date'], index_col='Date') ## parse_date is useful to convert dates to a single format
```

```python
appledf.head()
```

```python
appledf['Close/Last'] = appledf['Close/Last'].str.replace('$', '')
```

```python
appledf.head()
```

```python
appledf['Close/Last'] = appledf['Close/Last'].apply(lambda x: float(x))
```

```python
appledf['Close/Last'].plot()
```

```python
## asfreq() 
appledf = appledf.asfreq('H', method='pad')
```

```python
## I can create a dates column if I know the starting and ending date.
firtsDate = appledf.index.min()
lastDate = appledf.index.max()
dateRange = pd.date_range(appledf.index.min(),  appledf.index.max(), freq='B') ## 'B' returns only business days
appledf['date2'] = pd.date_range(dateRange.min(),  dateRange.max(), freq='H')
```

```python
## I can reset the index like this to a default value
appledf = appledf.reset_index(drop=True) ## this will DELETE the index column because of drop=True
appledf.head()
```

```python
## I now assigned the date2 column as new index column
appledf.set_index(appledf['date2'], inplace=True)
appledf.head()
```

```python
## SUBSTRACTING DIFFERENCES BETWEEN DATE SERIES with DIFFERENCE() method CHECKKKKKKKKkkk
range1 = appledf['date2']
range2 = pd.date_range('2022-05-01', '2023-01-01', freq='H')
# display(range3) ## 5881 rows
# display(range1) ## 8737 rows
rng = range2.difference(range1)
rng
```

```python
## CALCULATIONG AVG PRICE CLOSE PER MONTH (HISTORICAL, AVG CALCULATED FOR ALL YEARS)
appledf['month'] = appledf['date2'].apply(lambda x: x.month)
appledf.groupby(['month'])['Close/Last'].mean()
```

```python
## this allows to group data by time intervals
## the DataFrame must have a datetime-like index
appledf['Close/Last'].resample("M").mean()
```

```python
############ WORKING WITH SQL ########################
import sqlite3 ## mysql db build-in in python by default   
```

```python
# Connect to database
# connect() creates a database
# takes the path to the database file as an argument. If the
# file does not exist, the sqlite3 module will create an empty database. 
conn = sqlite3.connect('lucasdb.db')

#create a curser object
# The cursor is what you
# use to send SQL commands to your database via its execute() function. 
c = conn.cursor()

#create a new table
c.execute('''CREATE TABLE STOCKS(DATE TEXT,QTY REAL, PRICE REAL)''')


#INSERT SOME DATA
c.execute("INSERT INTO STOCKS VALUES('2023-03-05',5,35.14)")
c.execute("INSERT INTO STOCKS VALUES('2023-03-06',100,36.14)")
conn.commit()

c.close()
conn.close()
```

```python
conn = sqlite3.connect('lucasdb.db')

c =conn.cursor()
c.execute('SELECT * FROM STOCKS')

rows= c.fetchall()

for row in rows:
    print(row)

c.close()
conn.close()
```
